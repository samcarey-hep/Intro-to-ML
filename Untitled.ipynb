```python
import numpy as np
import pandas as pd
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import math
import matplotlib as mt
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, f1_score, recall_score
import seaborn as sns
import statistics
sns.set(style="whitegrid")
```

```python
dp = pd.read_csv("adult.csv")
df=dp.replace('?', np.NaN)
df
```

```python
def normalize(X, mean, std):
    return (X - mean) / std
```

```python
# Categorise the attributes
continuous_variables  = ['age', 'fnlwgt', 'education.num', 'capital.gain', 'capital.loss', 'hours.per.week']
categorical_variables = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country', 'income']
```

```python
#scaling continuous variables
def scaleFeature(df, col):
    mean = df[col].mean()
    std = df[col].std()
    df[col] = df[col].apply(lambda x: normalize(x, mean, std))
```

```python
oneh_encoded_data = pd.get_dummies(df, columns = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country', 'income'])
data = oneh_encoded_data
```

```python
for col in continuous_variables:
    scaleFeature(data, col)
```

```python
data = data.drop('income_>50K', axis=1)
```
